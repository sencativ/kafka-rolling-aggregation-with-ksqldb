```python
import json
import uuid
import os
import json
from dotenv import load_dotenv
from pathlib import Path
from kafka import KafkaProducer
from faker import Faker
from time import sleep
```

```python
dotenv_path = Path('/resources/.env')
load_dotenv(dotenv_path=dotenv_path)
```

```python
kafka_host = os.getenv('KAFKA_HOST')
# this kafka topic has 3 partition which i declare on the kafka ui
kafka_topic = "assignment-reza"
```

```python
from kafka import KafkaConsumer

cumulative_sum = 0
message_count = 0

# To consume latest messages and auto-commit offsets
consumer = KafkaConsumer(
    kafka_topic,
    group_id='dibimbing-group',
    bootstrap_servers=[f'{kafka_host}:9092'],
    value_deserializer=lambda m: json.loads(m.decode('utf-8')),  # Deserializer JSON
    auto_offset_reset='earliest'
)
```

```python
for message in consumer:
    record = message.value  # Get the message value
    
    # Extract the value from message to variable
    transaction_id = record['transaction_id']
    user_name = record['user_name']
    crypto_type = record['crypto_type']
    transaction_amount = record['transaction_amount']
    transaction_fee = record['transaction_fee']
    timestamp = record['timestamp']
    
    # Update cumulative sum and message count every new message
    cumulative_sum += transaction_amount
    message_count += 1
    
    # Calculate cumulative average
    cumulative_avg = cumulative_sum / message_count if message_count > 0 else 0
    
    # Print all needs
    print(f"Transaction ID: {transaction_id}")
    print(f"User Name: {user_name}")
    print(f"Crypto Type: {crypto_type}")
    print(f"Transaction Amount: {transaction_amount}")
    print(f"Transaction Fee: {transaction_fee}")
    print(f"Timestamp: {timestamp}")
    # Print the rolling aggregation
    print(f"Cumulative Sum: {cumulative_sum}")
    print(f"Cumulative Average: {cumulative_avg}")
    print("-" * 40)  # Separator for readability
```
